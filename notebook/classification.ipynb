```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
pd.options.display.float_format = '{:.1f}'.format 
# 숫자가 길어서 짤림

data = pd.read_csv("data_in_sample.csv", index_col=False)
data.head()
```

```python
print(data.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaled_data = MinMaxScaler().fit_transform(data.values.astype(float))
scaled_data = pd.DataFrame(scaled_data, columns=data.columns)
scaled_data.head()
```

```python
scaled_data.describe()
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3, algorithm='auto')
model.fit(scaled_data)
predict = pd.DataFrame(model.predict(scaled_data))
predict.columns = ['predict']

result = pd.DataFrame(pd.concat([scaled_data, predict], axis=1)) 

result.head()
```

```python
print(result.groupby("predict").size())
```

```python
print(result.describe())
```

```python
features = list(result.columns)

print(features)
```

```python
features.remove("predict")

print(features)
```

```python
x = result[features]
y = result["predict"]

x.describe()
```

```python
y.describe()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=0,
                                                    test_size=0.2)

X_train.describe()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
print('Accuracy of K-NN classifier on training set: {:.2f}'.format(
    knn.score(X_train, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'.format(
    knn.score(X_test, y_test)))
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, y_train)
print('Accuracy of GNB classifier on training set: {:.2f}'
      .format(gnb.score(X_train, y_train)))
print('Accuracy of GNB classifier on test set: {:.2f}'
      .format(gnb.score(X_test, y_test)))
```

```python
from sklearn.svm import SVC

svm = SVC()
svm.fit(X_train, y_train)
print('Accuracy of SVM classifier on training set: {:.2f}'
      .format(svm.score(X_train, y_train)))
print('Accuracy of SVM classifier on test set: {:.2f}'
      .format(svm.score(X_test, y_test)))
```
